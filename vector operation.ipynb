add two vector

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,4,-5])

sum = v1+v2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='red')
plt.quiver(0,0,4,-5,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,6,-2,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

subtract of two vector

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,4,-5])

sub = v1-v2
print(sub)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='red')
plt.quiver(0,0,4,-5,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-2,8,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

multiply with vector

```python
v1 = np.asarray([0,0,2,3])
v2 = 2

mul = v1*v2
print(mul)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot product of two *Vector*

```python
v1 = np.array([2,3])
v2 = np.array([4,-5])

Dot = np.dot(v1,v2)
print(Dot)
```

Cross product of two vector

```python
v1 = np.array([2,3,6])
v2 = np.array([4,-5,8])

Cross = np.cross(v1,v2)
print(Cross)
```

```python
a = np.array([2, 3])
v = np.array([4, -5])
dot = np.dot(a,v)
mag = np.sqrt(sum(v**2))
proj = (dot/mag**2)*v
print(proj)
```
