if elif else statement

```python
a=10
b=20
c=15

if (b<a>c):
  print('a is greater')
elif (a<b>c):
  print('b is greater')
else:
  print('c is greater')
```

Nested if atatement

```python
a=20
b=15
c=10

if (a>b):
  if (a>c):
    print('a is greater')
  else:
    print('c is greater')
else:
  if (b>c):
    print('b is greater')
  else:
    print('c is greater')
```
