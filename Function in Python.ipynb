```python
#normal way
number= int(input("enter a number: "))

factorial =1

if number==0:
  print("factorial is 1")
else:
  for i in range(1, number+1):
    factorial=factorial*i
  print(factorial)  
```

```python
#creating function
def factorial_value(num):
  factorial=1

  if num == 0:
   return factorial
  else:
   for i in range(1,num+1):
     factorial=factorial*i
   return factorial  
   


```

```python
print(factorial_value(5))
```

```python
print(factorial_value(10))
```
