# float

```python
a=2.6
print(a)
type(a)

num= float(input("give a number: "))
print(num)


# convert float to int

b=3.4
c=int(b)
print(c)
type(c)
```

# complex

# Boolean

```python
x= 7>3
print(x)
type(x)
```

# Tuple

```python
T=(1,2,3,4,5,6, "kabir", False)
print(T)
print(T[3])
type(T)

```

# list

```python
list = [1,2,3,4,5,"kabir",True]
print(list)
type(list)

list.append(6)
print(list)

del list[2]
print(list)
print(list[4])
type(list)

#convert list to Tuple
x=tuple(list)
print(x)
type(x)

# convert list to set
y=set(list)
print(y)
type(y)
```

# Set

```python
# set does not allow duplicate value
s={1,2,3,4,5,4,3,2,6,7,5}
print(s)
type(s)

# cant convert into list
```

# Dictionary

```python
D={"name":"kabir", "city":"dhaka","age":23}
print(D)
print(D['name'])
print(D['age'])
print(D['city'])

# dictionary does not allow dupicate value
```
